```python
import pandas as pd
import os
```

```python
df = pd.DataFrame()
for filename in os.listdir("..\Colecciones"):
    if filename.endswith(".xlsx"):
        print(filename)
        df2 = pd.read_excel(f"..\Colecciones\{filename}")
    df = pd.concat([df,df2])
df.to_csv("..\Colecciones\MTG_Database.txt", header=None, index=None)
```
